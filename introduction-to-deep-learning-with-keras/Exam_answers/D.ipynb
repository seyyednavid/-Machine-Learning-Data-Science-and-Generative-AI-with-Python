```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Load dataset
data = pd.read_csv("/content/drive/MyDrive/ML/concrete_data.csv")
```

```python
# Split features and target
X = data.drop(columns=["Strength"])
y = data["Strength"]
```

```python
# Normalize the predictors manually using the formula
X_normalized = (X - X.mean()) / X.std()
```

```python
print(X_normalized)
```

```python
# Initialize list to store MSE values
mse_list = []

# Repeat 50 times
for _ in range(50):
    # 1. Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=np.random.randint(0, 1000))

    # 2. Build the model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))


    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=10, verbose=2)

    # 3. Evaluate the model
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    # Store MSE
    mse_list.append(mse)
```

```python
# 4. Calculate mean and standard deviation of MSE
mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)
```

```python
# Print results after normalization with 50 epochs and 3 hiddens layer
print(f"Mean of MSE over 50 runs: {mean_mse}")
print(f"Standard Deviation of MSE over 50 runs: {std_mse}")
```

```python
# Print results after normalization with 50 epochs and a hidden layer
print(f"Mean of MSE over 50 runs: {mean_mse}")
print(f"Standard Deviation of MSE over 50 runs: {std_mse}")
```

By increasing the number of hidden layers (from 1 to 3), the Mean MSE significantly decreased (130.316 → 87.840), showing that the model's predictions became more accurate on average.

**However**, the Standard Deviation of MSE increased (11.836 → 25.615), indicating that the performance became less consistent across the 50 runs.
